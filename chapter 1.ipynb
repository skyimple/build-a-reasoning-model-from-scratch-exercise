## 1.1 Defining reasionging in the context of LLMs  
what is LLM-base reasoning?  
In the context of LLMs, reasoning means the model shows how it got to an answer by explaining the intermediate steps before giving the final response  
Chain-of-Thought(CoT):In this process, the model
generates a series of intermediate steps or explanations that show how it arrives at its final answer  
it is important to emphasize that LLM reasoning differs from traditional, deterministic reasoning  


## 1.2 Understanding the standard LLM training pipeline  
two stage: pre-training and post-training  
The pre-training objective (goal) for the LLM is to learn to predict the next word (i.e.,
token) in these texts  
TOKEN：A token is a small unit of text that a language model processes. A token can be a full word, part of a
word, or even punctuation, depending on how the text is split by a so-called tokenizer.  
pre-trained models merely serve as base models for the post-training stage, which uses two key techniques:
fine-tuning(SFT, also instruction tuning) and preference tuning(Reinforcement Learning with Human Feedback)  

## 1.3 Imporving LLM reasoning with training and inference techniques  
three approaches commonly used to improve reasoning capabilities in LLMs  
### 1.3.1 inference-compute scaling  
inference compute scaling or test-time scaling  includes methods that improve model reasoning capabilities at inference time (when a user prompts the model) <span style="color:red;">without training or modifying the underlying model weights</span>  
### 1.3.2 Reinforcement learning   
Reinforcement learning (RL) refers to training methods, so we need to train the model;  
<span style="color:red;">**RL updates the model's weights during training**</span>;his enables the model to learn and refine reasoning strategies through trial and error  
RLHF(preference tuning depends on human/ reasoning RL depends on system itself's reward)
### 1.3.3 Distillation   
Within the context of LLMs, this typically means performing
supervised fine-tuning (SFT) using high-quality labeled instruction datasets generated by a larger,
